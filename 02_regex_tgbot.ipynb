```python
with open('MkGl(Li).txt', 'r', encoding='utf-8') as file:
  MkLi = file.read()
with open('MkGl(Ru).txt', 'r', encoding='utf-8') as file:
  MkRu = file.read()
with open('LkGl (Li).txt', 'r', encoding='utf-8') as file:
  LkLi = file.read()
with open('LkGl (Ru).txt', 'r', encoding='utf-8') as file:
  LkRu = file.read()
with open('JnGl (Li).txt', 'r', encoding='utf-8') as file:
  JnLi = file.read()
with open('JnGl (Ru).txt', 'r', encoding='utf-8') as file:
  JnRu = file.read()



import nest_asyncio
nest_asyncio.apply()

from telegram import Update
from telegram.ext import ApplicationBuilder, CommandHandler, MessageHandler, ContextTypes, filters
import pandas as pd
import re
import io

def build_ref_dict(lines):
    ref_dict = {}
    for line in lines:
        match = re.search(r'\((\d+\.\d+)\)', line)
        if match:
            ref = match.group(1)
            ref_dict[ref] = line.strip()
    return ref_dict

async def handle_message(update: Update, context: ContextTypes.DEFAULT_TYPE):
    user_input = update.message.text.strip()
    pattern = re.compile(rf'\b{re.escape(user_input)}[\w]*\b', re.IGNORECASE)

    gospels = [
        ("Mk", MkLi.strip().split('\n'), MkRu.strip().split('\n')),
        ("Lk", LkLi.strip().split('\n'), LkRu.strip().split('\n')),
        ("Jn", JnLi.strip().split('\n'), JnRu.strip().split('\n')),

    ]

    all_rows = []

    for gospel_name, li_lines, ru_lines in gospels:
        li_dict = build_ref_dict(li_lines)
        ru_dict = build_ref_dict(ru_lines)

        all_refs = sorted(set(li_dict.keys()) | set(ru_dict.keys()))

        for ref in all_refs:
            li_line = li_dict.get(ref, "")
            ru_line = ru_dict.get(ref, "")

            match_li = pattern.search(li_line)
            match_ru = pattern.search(ru_line)

            if match_li or match_ru:
                form_li = match_li.group() if match_li else ""
                form_ru = match_ru.group() if match_ru else ""

                all_rows.append({
                    'Gospel': gospel_name,
                    'Verse': ref,
                    'Context Li': li_line,
                    'Form Li': form_li,
                    'Form Ru': form_ru,
                    'Context Ru': ru_line,
                })

    df = pd.DataFrame(all_rows)
    pd.set_option('display.max_colwidth', None)

    if df.empty:
        await update.message.reply_text("Nothing found.")
    else:
        buffer = io.BytesIO()
        df.to_excel(buffer, index=False, sheet_name="Results")
        buffer.seek(0)
        await update.message.reply_document(document=buffer, filename="results.xlsx")







async def start(update: Update, context: ContextTypes.DEFAULT_TYPE):
    await update.message.reply_text("Ƿes þū hāl! Send me a word!")

async def stop(update: Update, context: ContextTypes.DEFAULT_TYPE):
    await update.message.reply_text("Stopping the bot...Far wel!")
    raise SystemExit

def main():
    app = ApplicationBuilder().token("7998108568:AAHuJ1SSvBeqkmNAvFTeSNbIQhMQMKuj3Qk").build()

    app.add_handler(CommandHandler("start", start))
    app.add_handler(CommandHandler("stop", stop))
    app.add_handler(MessageHandler(filters.TEXT & ~filters.COMMAND, handle_message))

    print("Bot is launched.")
    app.run_polling()

if __name__ == "__main__":
    main()
```
